```
from pandas_datareader import data as wb
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA
```

```
data = wb.get_data_yahoo('BTC-USD', start='2015-1-1', end='2019-11-1')
data.tail()
```

```
class Smacross(Strategy):
    def init(self):
        Close = self.data['Adj Close']
        self.m1 = self.I(SMA, Close, 9)
        self.m2 = self.I(SMA, Close, 21)

    def next(self):
        if crossover(self.m1, self.m2):
            self.buy()
        
        elif crossover(self.m2, self.m1):
            self.sell()
```

```
bt = Backtest(data, Smacross, cash=1000, commission=0.002)
```

```
bt.run()
```

```
bt.plot()
```
